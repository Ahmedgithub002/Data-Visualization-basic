### Importing Libraries

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

To install Matplotlib in your local machine, you need to use the command

`pip install matplotlib`

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Introduction to Matplotlib

```python
# .plot() function
x = [1,4,3,5]
y = [4,8,9,6]

# plot
plt.plot(x,y)
plt.show()
```

```python
# Adding Multiple Lines in .plot() funtion
plt.plot([1,4,3,7],[2,4,6,8])
plt.plot([2,3,4,5],[1,3,5,7])
plt.plot([4,3,2,1],[1,2,3,4])

plt.show()
# To show different graphs for each data, you need to create subplots
```

```python
# Adding Data With Numpy
x = np.arange(1, 5)

# x -> array([1, 2, 3, 4])

plt.plot(x,x)
plt.plot(x,x+1)
plt.plot(x,x+2)
plt.show()
```

### Customising Graphs

```python
# Adding Title, X and Y Axis Labels
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

# Title of Graph
plt.title("Title of th graph")
# Labels
plt.xlabel("X Axis")
plt.ylabel("y Axis")
plt.show()
```

#### Legend, Grid and Axes

LEGEND: for location of legend available options are: 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

AXIS: Axis command takes 4 values [xstart,xfinal,ystart,yfinal]

```python
# Adding Legend
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
plt.legend(["One","Two","Three"])
# It takes array as input for each line in plot and assigns the title to it


plt.show()
```

```python
# Adding Legend in Plot function itself
x = np.arange(1, 5)

# CODE HERE
plt.plot(x, x, label="One")
plt.plot(x, x+1,label="Two")
plt.plot(x, x+2,label="Three")

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
# It takes array as input for each line in plot and assigns the title to it
plt.legend(loc='best')

plt.show()
```

```python
# Enabling Grid and Axis Modification
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
# It takes array as input for each line in plot and assigns the title to it
plt.legend(['One', 'Two', 'Three'], loc='best')

# Grid
plt.grid(True)
# Axes
plt.axis([0,7,2,8])
plt.show()
```

#### Line and Marker Customisation

Available Shapes for Markers are:

`"s"=square, "p"=pentagon, "."=point, "o"=circle, "d"=diamond, etc.`

Available Lines are:

`"-", ":", "--", "-.", etc.`

<br>
By Default, third argument in plot() function is line pattern alongside other customizations like `color, markersize, linestyle, linewidth, marker,
 markeredgecolor,  etc`

<br>
 More can be read at:

 LineStyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

 MarkerStyle: https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html

```python
# Line and Marker Styling
x = np.arange(1, 5)

# STYLE your PLOTS
plt.plot(x, x, color='blue',linestyle=':',linewidth=3.0, marker='d')
plt.plot(x, x+1,color='pink',linestyle='-',linewidth=6.0, marker='p')
plt.plot(x, x+2,color='yellow',linestyle='-.',linewidth=4.0, marker='o')

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
# It takes array as input for each line in plot and assigns the title to it
plt.legend(['One', 'Two', 'Three'], loc='best')

plt.show()
```

### Subplots to create multiple plots

```python
# Subplot

# Create a figure
fig=plt.figure(figsize=(16,8), facecolor='cyan')
# subplot syntax => fig.add_subplot(rows, columns, poisition)
rows =2
columns=3
# Follows a grid format => rows x columns
# Create charts on figure using add_subplot
chart1=fig.add_subplot(rows,columns,1)
chart2=fig.add_subplot(rows,columns,2)
chart3=fig.add_subplot(rows,columns,3)
chart4=fig.add_subplot(rows,columns,4)

# Each of the above subplots act as separate plot and can be used to plot anything in space
x = np.arange(1, 5)
chart1.plot(x,x,":")
chart2.plot(x,x+1,"p")
chart3.plot(x,x+2,".:")
chart4.plot(x,x+3,"-")

# Plot on subplots

plt.show()
```

### Pie-Chart and Exploded Pie-Chart

```python
plt.figure(figsize=(6,6))

# labels and values
labels =["India", "USA", "Britain", "Germany", "Japan"]
values =[25, 30, 20, 10, 35]

# Plot pie chart without values and with values
plt.pie(values,labels=labels,autopct='%1.1f%%')

plt.show()
```

Explode is an array which specifies the fraction of the radius with which to offset each wedge.

```python
plt.figure(figsize=(6,6))

labels = ["India", "USA", "Britain", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]

# To create exploding pie-chart, define an exploding amount array
exp =[0,0.1,0.3,0.4,0.05]

# add explode in pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%',explode=exp)
plt.show()
```

### Bar-Charts

```python
plt.figure(figsize=(7,5), facecolor='cyan')

# labels and values for bar chart
labels = ["India", "USA", "Britain", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]

# Plot the bar chart
plt.bar(labels,values)
plt.show()
```

Grouped Bar Charts

```python
plt.figure(figsize=(7,5), facecolor='cyan')

# labels and values for grouped chart
labels = ["India", "USA", "Britain", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]
values2 = [30, 20, 15, 15, 40]

# Specify Label Positions for tickers on x-axis
positions = np.arange(len(labels))

# draw 2 bar charts for displaying grouped bars alongside widths
plt.bar(positions,values,color='red',width=0.3)
plt.bar(positions+0.3,values,color='blue',width=0.3)

# give position to display x tickers on X Axis, which should be position of bar + half of width of bars


plt.show()
```

### Scatter Plots

```python
# specify points locations
x = np.random.randn(200)
y = np.random.randn(200)

# give size and color array for the scatter plot
size = 100*np.random.randn(200)
colors = np.random.randn(200)

# plot the scatter plot
plt.scatter(x,y,s=size,c=colors,marker="p")

plt.show()
```

### Introduction to 3D Line Plots

```python
# directions
x = [5,9,1,3,7]
y = [6,8,2,4,10]
z = [6,1, 4,1,9]

# adding 3d projection by creating a subplot to plot the line graph
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot(x,y,z)

plt.show()
```

### 3D Scatter Plots

```python
# give positions for graphs
x = np.random.randn(200)
y = np.random.randn(200)
z = np.random.randn(200)

size = 100*np.random.randn(200)
colors = np.random.randn(200)

# create a 3d projection subplot to display our scatter plot
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c=colors,s=size,marker='^')

plt.show()
```

### 3D Bar Plots

```python
# create a 3d projection subplot to display our scatter plot
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
# define distances from plane
x = [1,2,3,4,5]
y = [3,6,2,8,5]
z = [0,0,0,0,0]
# length of bar in direction plane
dx =[1,1,1,1,2]
dy =[1,1,2,1,1]
dz =[1,8,1,4,2]

# plot the 3d bar graph using both distances and lengths of bar
ax.bar3d(x,y,x, dx,dy,dz, color='gold')

plt.show()
```
